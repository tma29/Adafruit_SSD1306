```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problem 1 with random dataset of 500 classes 
#Then I took 10% and 15% of the total 1000 samples, split up again. 
```

```python
x = np.random.multivariate_normal([-1, -1], [[1, -.25], [-.25, 1]], 500).T
y = np.random.multivariate_normal([1, 1], [[1, -.25], [-.25, 1]], 500).T
plt.plot(x[0], x[1], 'o', c='b')
plt.plot(y[0], y[1], 'o', c='r')
plt.axis('equal')
plt.xlabel('$x_1$')  # use latex in the figure axis labels
plt.ylabel('$x_2$')
plt.legend(("Class 1", "Class 2"))
plt.show()
```

```python
#taking 10% of the data 
x = np.random.multivariate_normal([-1, -1], [[1, -.25], [-.25, 1]], 450).T
y = np.random.multivariate_normal([1, 1], [[1, -.25], [-.25, 1]], 450).T
plt.plot(x[0], x[1], 'o', c='b')
plt.plot(y[0], y[1], 'o', c='r')
plt.axis('equal')
plt.xlabel('$x_1$')  # use latex in the figure axis labels
plt.ylabel('$x_2$')
plt.legend(("Class 1", "Class 2"))
plt.show()
```
